```python
# Paste this into a new Jupyter notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

ROOT = Path(r"E:\Labmentix\projects\EMIpredict")
df = pd.read_parquet(ROOT / "data" / "processed" / "emi_clean.parquet")
```

```python
# 1. Eligibility Distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='emi_eligibility', order=['Eligible', 'High_Risk', 'Not_Eligible'])
plt.title('EMI Eligibility Distribution')
plt.show()

print(df['emi_eligibility'].value_counts(normalize=True))
```

```python
# 2. By EMI Scenario
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='emi_scenario', hue='emi_eligibility')
plt.xticks(rotation=45)
plt.title('Eligibility by EMI Scenario')
plt.show()
```

```python
# 3. DTI Ratio
df['dti'] = (df['current_emi_amount'] + df['monthly_rent'] + df['school_fees'] + 
             df['college_fees'] + df['groceries_utilities'] + df['other_monthly_expenses']) / df['monthly_salary']

sns.boxplot(
    data=df, 
    x='emi_eligibility', 
    y='dti',
    palette='viridis',     
    width=0.9,             
    linewidth=1.5,         
    notch=True,            
    fliersize=8,           
    showfliers=True        
)

plt.ylim(0, 3.0) 

plt.title('Debt-to-Income Ratio by Eligibility')
plt.ylabel('Debt-to-Income (DTI) Ratio')
plt.show()
```

```python
# 4. Correlation Heatmap
num_cols = df.select_dtypes('number').columns
sns.heatmap(df[num_cols].corr(), cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```
